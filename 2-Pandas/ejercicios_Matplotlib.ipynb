6. Usando una figura Y EJES POR SEPARADO, usa numpy para los valores del eje X entre 0 y 5. Pinta dos gr치ficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). Elige la precisi칩n en el eje X que desees.  

```python
x = np.linspace(0,5, 100)
y1 = 3*x+0
y2 = -3*x-5
plt.subplot(1, 2, 1)
plt.plot(x,y1)
plt.ylim(0, 15)
plt.xlim(0, 5)
plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.ylim(-7, -20)
plt.xlim(0, 5)
```

```python

# Crear valores de ejes X entre 0 y 5
x_values = np.linspace(0, 5, 100)

# Calcular los valores de las rectas
y1_values = 3 * x_values  # Pendiente positiva de 3
y2_values = -3 * x_values - 5  # Pendiente negativa de 3, pasando por (0, -5)

# Crear figura y ejes separados
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Graficar la recta con pendiente positiva en el primer subplot
axs[0].plot(x_values, y1_values, color='blue')
axs[0].set_title('Recta con pendiente positiva (m=3)')
axs[0].set_xlabel('Eje X')
axs[0].set_ylabel('Eje Y')
axs[0].grid(True)

# Graficar la recta con pendiente negativa en el segundo subplot
axs[1].plot(x_values, y2_values, color='red')
axs[1].set_title('Recta con pendiente negativa (m=-3)')
axs[1].set_xlabel('Eje X')
axs[1].set_ylabel('Eje Y')
axs[1].grid(True)

# Ajustar el espaciado entre subplots
plt.tight_layout()

# Mostrar la gr치fica
plt.show()
```
