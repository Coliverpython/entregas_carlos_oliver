```python
# 23. Mapas de calor
# Calcula las correlaciones del primer DataFrame original df
# Pista: las correlaciones se calculan con el método corr()
# Recuerda: las correlaciones miden entre dos variables, si varío una, cuánto me varía la otra (cómo están de relacionadas)
```

```python
df_numeric = pokemon.select_dtypes(include=['float64', 'int64'])
corr = df_numeric.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.05)
plt.title('Mapa de calor de correlación')
plt.show()
```

```python
# 27. Muestra una función distribución de probabilidad con kdeplot (curvas de nivel) de la defensa respecto al ataque.
# Di un bin 2D donde caigan muchos puntos (ataque, defensa). joinplot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un kdeplot para la distribución de probabilidad conjunta
plt.figure(figsize=(8, 6))
sns.kdeplot(data=pokemon, x='Attack', y='Defense', fill=True, cmap='viridis')

# Agregar un joinplot para visualizar la distribución de probabilidad conjunta con histogramas marginales
sns.jointplot(data=pokemon, x='Attack', y='Defense', kind='hist', cmap='viridis', bins=20)

# Añadir título al gráfico kdeplot
plt.title('Distribución de Probabilidad Conjunta de Ataque y Defensa')

# Mostrar los gráficos
plt.show()
```
