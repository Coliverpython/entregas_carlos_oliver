# Occupation

```python
import pandas as pd
```

### Step 3. Assign it to a variable called users.

```python
users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                      sep='|', index_col='user_id')
users 
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby("occupation")[["age"]].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
filtro_m = users[users["gender"] == "M"]
filtro_m_counts = filtro_m["occupation"].value_counts()
users_counts = users["occupation"].value_counts()
percentage = filtro_m_counts.div(users_counts.reindex(filtro_m_counts.index), fill_value=0) * 100
percentage
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby("occupation")[["age"]].min()
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(["occupation","gender"])["age"].mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
total_por_ocupacion = users.groupby("occupation")["gender"].value_counts().sort_values()
porcentaje_por_ocupacion = total_por_ocupacion.groupby(level=0).apply(lambda x: 100 * x / float(x.sum()))
porcentaje_por_ocupacion
```
