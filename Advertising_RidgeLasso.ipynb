## Modelado

```python
X = adver[['TV','radio', 'newspaper']]#tv, radio y newspaper
y = adver['sales'] #sales
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state= 0)
```

```python
#Miramos los shape de cada test y train para ver si lo hemos hecho bien

print(X_train.shape) 
print(X_test.shape) 
print(y_train.shape) 
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
from sklearn import linear_model, metrics, model_selection
from sklearn.linear_model import ElasticNet
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
```

```python
#instancia y fit del modelo lineal
from sklearn import linear_model, metrics, model_selection

lr = linear_model.LinearRegression()

lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)

base_line_error =  metrics.mean_squared_error(y_test,y_pred)


print('MAE test', mean_absolute_error(y_test, y_pred))
print('MAPE test', mean_absolute_percentage_error(y_test, y_pred))
print('MSE test', mean_squared_error(y_test, y_pred))
print('RMSE test', np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
# instancia y fit del ridge
from sklearn.linear_model import Ridge

ridgeR = Ridge(alpha = 10)
ridgeR.fit(X_train, y_train)


print('MAE test', mean_absolute_error(y_test, ridgeR.predict(X_test)))
print('MSE test', mean_squared_error(y_test, ridgeR.predict(X_test)))
print('RMSE test', np.sqrt(mean_squared_error(y_test, ridgeR.predict(X_test))))

```

```python
n_alphas = 40
alphas = np.logspace(1, 5, n_alphas) 
alphas
```

```python
n_alphas = 40
alphas = np.logspace(1, 5, n_alphas) 
alphas
```

```python

coef_ridge = []
err_ridge = []
baseline = []

errors = {}

for a in alphas:
    ridge = Ridge(alpha=a)
    ridge.fit(X_train, y_train)
    
    coef_ridge.append(ridge.coef_)
    
    y_pred = ridge.predict(X_test)
    ridge_error = metrics.mean_squared_error(y_pred, y_test)
    
    err_ridge.append(ridge_error)
    baseline.append(base_line_error)
```

```python
# alpha buena
alphas[err_ridge.index(min(err_ridge))]
```

```python
#Soms demasiado buenos y esta grafica os la dejamos.
plt.figure(figsize=(20,12))
ax = plt.gca()
ax.plot(alphas, err_ridge, linewidth=5, color='red', label="Ridge regression")
ax.plot(alphas, baseline, linewidth=4,linestyle='--', color='blue', label='Linear regression')
ax.set_xscale('log')
plt.xlabel('$\lambda$', fontsize=30)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.ylabel('error', fontsize=30)
ax.legend(fontsize=30)
plt.title(r'Regression error ($\lambda$)', fontsize=30)
plt.ylim(None, 4500000)
plt.show()
```

```python
n_alphas = 100
alphas = np.logspace(1, 4, n_alphas) 

coef_lasso = []
err_lasso = []
baseline = []

for a in alphas:
    # de nuevo buscamos alpha con el lasso
    lasso = linear_model.Lasso(alpha=a)    
    lasso.fit(X_train, y_train)
    coef_lasso.append(lasso.coef_)
    y_pred = lasso.predict(X_test)
    lasso_error = metrics.mean_squared_error(y_pred, y_test)    
    err_lasso.append(lasso_error)
```

```python
baseline = [0] * len(alphas)
```

```python
import matplotlib.pyplot as plt

# Trazar los resultados
plt.figure(figsize=(10, 6))

# Coeficientes Lasso
plt.subplot(1, 2, 1)
plt.semilogx(alphas, coef_lasso)
plt.xlabel('Alpha')
plt.ylabel('Coeficientes')
plt.title('Coeficientes Lasso')

# Errores Lasso
plt.subplot(1, 2, 2)
plt.semilogx(alphas, err_lasso)
plt.xlabel('Alpha')
plt.ylabel('Error cuadrático medio (MSE)')
plt.title('Errores Lasso')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 12))
ax = plt.gca()

# Trazar el error de regresión para el modelo Lasso
ax.plot(alphas, err_lasso, linewidth=5, color='red', label="Lasso Regression")

# Trazar el error de referencia (baseline)
# Asegúrate de que 'baseline' esté definido y tenga la misma longitud que 'alphas'
ax.plot(alphas, baseline, linewidth=4, linestyle='--', color='blue', label='Linear Regression')

ax.set_xscale('log')
plt.xlabel('$\lambda$', fontsize=30)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.ylabel('Error', fontsize=30)
ax.legend(fontsize=30)
plt.title(r'Regression Error ($\lambda$)', fontsize=30)
plt.show()
```
