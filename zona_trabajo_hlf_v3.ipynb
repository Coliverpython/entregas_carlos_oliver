```python

class HundirLaFlota:
    def __init__(self ,filas=10, columnas=10, vidas_pc = 4, vidas_jg = 4):
        """
        Inicializa el juego con el tamaño del tablero especificado.
        """
        self.filas = filas
        self.columnas = columnas
        self.tablero_jugador = np.full((self.filas, self.columnas), " ")
        self.tablero_ordenador = np.full((self.filas, self.columnas), " ")
        self.barcos = barcos = {
                                "Acorazado": {"posiciones": [], "golpes": 0, "tamaño": 5},
                                "Crucero": {"posiciones": [], "golpes": 0, "tamaño": 4},
                                "Buque": {"posiciones": [], "golpes": 0, "tamaño": 3},
                                "Yate": {"posiciones": [], "golpes": 0, "tamaño": 2},   
                                }
        self.posiciones_barcos_jg = []
        self.posiciones_barcos_ordenador = []


    def colocar_barcos_aleatorios(self,tablero, nombre_barco, jugador = False, n_barcos = 2):
        """
        Coloca los barcos aleatoriamente en el tablero.
        """
        size = self.barcos[nombre_barco]["tamaño"]

        while n_barcos >= 1:
            # Genera posiciones aleatorias en el tablero
            fila = np.random.randint(0, tablero.shape[0])
            columna = np.random.randint(0, tablero.shape[1])
            orientacion = np.random.choice(["horizontal", "vertical"])
            # Escoge aleatoriamente el tamaño de un barco
            size = self.barcos[nombre_barco]["posicion"]
            if orientacion == "horizontal" and columna + size <= self.columnas:
                if self.verificar_adyacencia(tablero, fila, columna, size):
                    for i in range(size):
                        tablero[fila, columna + i] = "B"
                        self.barcos[nombre_barco]["posiciones"].append((fila, columna + i))
                    n_barcos -= 1

            elif orientacion == "vertical" and fila + size <= self.filas:
                if self.verificar_adyacencia(tablero, fila, columna, size):
                    for i in range(size):
                        tablero[fila + i, columna] = "B"
                        self.barcos[nombre_barco]["posiciones"].append((fila + i, columna))
                    n_barcos -= 1
                            
                        
                    
    def verificar_adyacencia(self, tablero, fila, columna, size):
        """
        Verifica si hay suficiente espacio alrededor de una posición para colocar un barco sin solaparse con otros barcos.
        """
        for i in range(-1, size + 1):
            for j in range(-1, size + 1):
                if 0 <= fila + i < tablero.shape[0] and 0 <= columna + j < tablero.shape[1]:
                    if tablero[fila + i, columna + j] == "B":
                        return False
        return True
    
    
    def disparo_simple(self, tablero, fila, columna):
        '''
        En la entrada de esta función hay que hacer referencia al tablero del contrincante.
        '''
        if tablero[fila, columna] == "B":
            tablero[fila, columna] = "X"
            print("Has tocado un barco")
            self.mostar_tableros()
            return True  # Indicar que el disparo fue exitoso
        if tablero[fila, columna] == "X" and tablero[fila, columna + 1]  == "X" and tablero[fila + 1, columna]  == "X" and tablero[fila, columna- 1] == "X" and tablero[fila - 1, columna] == "X":
            print("Barco Hundido")
        elif tablero[fila, columna] == " ":
            tablero[fila, columna] = "0"
            print("¡Agua!")
            self.mostar_tableros()
            return False  # Indicar que el disparo falló

        self.mostar_tableros()

    def turno_ordenador(self):
        while True:
            fila = np.random.randint(0, self.tablero_ordenador.shape[0])
            columna = np.random.randint(0, self.tablero_ordenador.shape[1])
            if self.disparo_simple(self.tablero_jugador, fila, columna):
                # Si el disparo fue exitoso, el ordenador repite su turno
                continue
            else:
                break
        

    def jugar(self):
        self.colocar_barcos_aleatorios(self.tablero_ordenador)
        self.colocar_barcos_aleatorios(self.tablero_jugador, jugador = True)
        self.mostar_tableros()
        
        while True:
            print("\n Tu turno")
            while True:
                fila = int(input("Fila a disparar?"))
                columna = int(input("Columna a disparar?"))
                if self.disparo_simple(self.tablero_ordenador, fila, columna):
                    # Si el disparo fue exitoso, repetir el turno del jugador
                    continue
                else:
                    break

            if not np.any(self.tablero_ordenador == "B"):
                print("Has ganado!")
                break

            print("\n Turno del ordenador")
            self.turno_ordenador()

            if not np.any(self.tablero_jugador == "B"):
                print("Has perdido, no te quedan más barcos")
                break
                



    def mostar_tableros(self):
        clear_output(wait=True)
        time.sleep(1)
        print(f"Tablero ordenador \n {self.tablero_ordenador}")
        print(f"\nTablero jugador \n {self.tablero_jugador}")
        print(f"Estas son las posiciones donde estan los barcos ordenador {self.posiciones_barcos_ordenador}")
        print(f"Estas son las posiciones donde estan los barcos jugador {self.posiciones_barcos_jg}")
        time.sleep(2)
        



```
