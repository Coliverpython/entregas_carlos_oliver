```python
import pandas as pd

# Definir la lista de tuplas
lista_alumnos = [("Caranchoa", 24, 1.82), 
                 ("Leo Messi", 25, 1.71), 
                 ("El bicho", 35, 1.66), 
                 ("Paquirrin", 37, 1.63), 
                 ("John Cobra", 29, 1.90), 
                 ("El tomate", 27, 1.80), 
                 ("El chucky de cieza", 28, 1.70), 
                 ("Chiquito de la Calzada", 32, 1.60), 
                 ("El cuñao", 36, 1.74), 
                 ("Rafa Nadal", 45, 1.72), 
                 ("Pau Gasol", 41, 1.65), 
                 ("El hermano de Pau Gasol", 30, 1.65), 
                 ("Joselito", 40, 1.60), 
                 ("Maria Isabel", 39, 1.80), 
                 ("Melody", 23, 1.77), 
                 ("Las ketchup", 32, 1.75), 
                 ("Pumuki", 27, 1.70), 
                 ("Naranjito", 34, 1.80), 
                 ("Espinete", 27, 1.77), 
                 ("El Feli", 28, 1.70), 
                 ("Jose Luis", 35, 1.80)]

# Convertir la lista de tuplas en un DataFrame de pandas
df_alumnos = pd.DataFrame(lista_alumnos, columns=["Nombre", "Edad", "Altura"])

# Mostrar el DataFrame
print(df_alumnos)
```

```python
# Divide el df en features (X) y target (y)
# Características (features)
X = df_alumnos[['Edad']]

# Objetivo (target)
y = df_alumnos[['Altura']]

```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 42)
# Divide los datos en conjunto de entrenamiento y test
# Siendo el tamaño del conjunto test un 30% y la semilla sea 42
```

```python
# Muestra el tamaño de X, X_train, y_train, X_test e y_test
print("Total features shape:", X.shape)
print("Train features shape:", X_train.shape)
print("Train target shape:", y_train.shape)
print("Test features shape:", X_test.shape)
print("Test target shape:", y_test.shape)
```

```python
# Instancia y entrena un regresor lineal
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm
```

```python
lm.fit(X_train, y_train)
```

```python
# imprime el intercepto y el coeficiente
print(lm.intercept_)
print(lm.coef_)
```

(Altura) = 1.85 - 0.0041*(Edad)

```python
# Realiza una predicción con el conjunto de test
predictions = lm.predict(X_test)
predictions
```

```python
# Crea un df con el conjunto de test
# Unir X_test y df_y_test
df_test = X_test.join(y_test)

# Mostrar el DataFrame resultante
df_test
```

```python
# Añade una columna con la predicción
df_test['Prediccion'] = predictions
df_test
```

```python
# Muestra el score del módelo
score = lm.score(X_test, y_test)
score
```

```python
# Muestra el Mean Absolute Error, el Mean Squared Error y el Root Mean Squared Error
mae = mean_absolute_error(df_test['Altura'], df_test['Prediccion'])
mse = mean_squared_error(df_test['Altura'], df_test['Prediccion'])
rmse = mean_absolute_percentage_error(df_test['Altura'], df_test['Prediccion'])
mae
mse
rmse
```

```python
import numpy as np
# Ahora vamos a calcular el MAE, MSE, RMSE y MAPE a mano

# Calcular las predicciones del modelo en el conjunto de prueba
predicciones = lm.predict(X_test)

# Calcular el Mean Absolute Error (MAE)
mae = np.mean(np.abs(predicciones - y_test))
print("Mean Absolute Error (MAE):", mae)

# Calcular el Mean Squared Error (MSE)
mse = np.mean((predicciones - y_test) ** 2)
print("Mean Squared Error (MSE):", mse)

# Calcular el Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print("Root Mean Squared Error (RMSE):", rmse)

# Calcular el Mean Absolute Percentage Error (MAPE)
mape = np.mean(np.abs((y_test - predicciones) / y_test)) * 100
print("Mean Absolute Percentage Error (MAPE):", mape)
```

```python
# Añade una columna con el error
# Calcular los errores entre las predicciones y los valores reales
errores = predicciones - y_test

# Agregar una columna 'Error' al DataFrame df_test
df_test['Error'] = errores

# Mostrar el DataFrame resultante
df_test
```
