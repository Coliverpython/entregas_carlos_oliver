```python
#7. Devolver los restaurantes que no preparan cocina americana y lograron un punto de calificación 'A' que no pertenece al distrito de Brooklyn. 
#El documento debe mostrarse según la cocina en orden descendente.
projection = {"_id": 0, "restaurant_id": 1, "name": 1, "cuisine": 1, "grades": 1, "address.coord": 1}  # Proyección para incluir solo los campos deseados
filter = {
    "cuisine": {"$ne": "American"},
    "grades.grade": "A",
    "address.coord.0": {"$ne": "Brooklyn"}
}  # Filtro para seleccionar los restaurantes que cumplen con las condiciones
sort = [("cuisine", -1)]  # Orden descendente por la cocina
result = list(restaurantes.find(filter=filter, projection=projection).sort(sort).limit(5))

for doc in result:
    print(doc)
```

```python
#8. Devolver los restaurantes que pertenecen al distrito Bronx y preparan platos americanos o chinos
projection = {"_id": 0, "restaurant_id": 1, "name": 1, "cuisine": 1, "grades": 1, "address.coord": 1}  # Proyección para incluir solo los campos deseados
filter = {
    "cuisine": {"$ne": "American"},
    "grades.grade": "A",
    "address.coord.0": {"$ne": "Brooklyn"}
}  # Filtro para seleccionar los restaurantes que cumplen con las condiciones
sort = [("cuisine", -1)]  # Orden descendente por la cocina
result = list(restaurantes.find(filter=filter, projection=projection).sort(sort).limit(5))

for doc in result:
    print(doc)
```

```python
#9. Devolver ID del restaurante, nombre, distrito y la cocina para aquellos restaurantes que pertenecen al distrito de Staten Island o Queens o Bronx o Brooklyn

projection = {"_id": 0, "restaurant_id": 1, "name": 1, "borough": 1, "cuisine": 1}  # Proyección para incluir solo los campos deseados
filter = {"borough": {"$in": ["Staten Island", "Queens", "Bronx", "Brooklyn"]}}  # Filtro para seleccionar los distritos deseados
result = list(restaurantes.find(filter=filter, projection=projection))

for doc in result:
    print(doc)
```

```python
projection = {"_id": 0, "name": 1, "borough": 1, "cuisine": 1}  # Proyección para incluir solo los campos deseados
filter = {"grades.score": {"$lte": 10}}  # Filtro para seleccionar restaurantes con puntaje no superior a 10
result = list(restaurantes.find(filter=filter, projection=projection))

for doc in result:
    print(doc)
```

```python
# 11. Devolver ID del restaurante, el nombre y las calificaciones del restaurante para aquellos restaurantes que obtuvieron una calificación de "A" y obtuvieron un puntaje de 11 en una fecha ISO "2014-08-11T00: 00: 00Z" entre muchas fechas de encuesta
projection = {"_id": 1, "name": 1, "grades": {"$elemMatch": {"date": {"$eq": ISODate("2014-08-11T00:00:00Z"), "grade": "A", "score": 11}}}}
filter = {"grades.date": ISODate("2014-08-11T00:00:00Z")}  # Filtro para seleccionar la fecha deseada
result = list(restaurantes.find(filter=filter, projection=projection).limit(5))

for doc in result:
    print(doc)
```

```python
# 12. Devolver ID del restaurante, nombre, dirección y ubicación geográfica del restaurante de aquellos donde el segundo elemento de la matriz coord contiene un valor que es más de 42 y hasta 52
projection = {"_id": 1, "name": 1, "address": 1, "coord": 1}  # Proyección para incluir solo los campos deseados
filter = {"address.coord.1": {"$gt": 42, "$lte": 52}}  # Filtro para seleccionar la coordenada deseada
result = list(restaurantes.find(filter=filter, projection=projection).limit(5))

for doc in result:
    print(doc)
```

```python

# 16. Borra los restaurantes con latitud menor que -95.754168
filter = {"address.coord.1": {"$lt": -95.754168}}  # Filtro para seleccionar los restaurantes con latitud menor que -95.754168
result = restaurantes.delete_many(filter).limit(5)

for doc in result:
    print(doc)
```

```python
# 17. Borra los restaurantes cuyo nombre empiece por 'C'
filter = {"name": {"$regex": "^C"}}  # Filtro para seleccionar los restaurantes cuyo nombre empiece por 'C'
result = restaurantes.delete_many(filter.limit(5))

for doc in result:
    print(doc)
```
