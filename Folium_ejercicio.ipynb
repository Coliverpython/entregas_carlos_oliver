```python
import folium
import json

map_geojson = folium.Map(location=[0, 0], zoom_start=3)

with open('russel_cook.json', 'r') as f:
    geojson_data = json.load(f)

# Inicializar listas para las coordenadas de los marcadores
green_marker_coords = []
blue_marker_coords = []

# Iterar sobre las características del GeoJSON
for feature in geojson_data['features']:
    # Obtener las coordenadas de la característica
    coordinates = feature['geometry']['coordinates']
    
    # Dividir las coordenadas en dos mitades
    midpoint = len(coordinates) // 2
    first_half = coordinates[:midpoint]
    second_half = coordinates[midpoint:]
    
    # Agregar marcadores a la primera mitad con el icono 'poo'
    for coord in first_half:
        location = tuple(coord[::-1])  # Revertir las coordenadas (geojson -> lat, lon)
        folium.Marker(location=location,
                      icon=folium.Icon(color='green', icon='poo', prefix='fa')).add_to(map_geojson)
        green_marker_coords.append(location)
    
    # Agregar marcadores a la segunda mitad con el icono 'star'
    for coord in second_half:
        location = tuple(coord[::-1])  # Revertir las coordenadas (geojson -> lat, lon)
        folium.Marker(location=location,
                      icon=folium.Icon(color='blue', icon='hippo', prefix='fa')).add_to(map_geojson)
        blue_marker_coords.append(location)

# Agregar una línea de ruta que conecte los marcadores
folium.PolyLine(locations=green_marker_coords + blue_marker_coords, color='blue').add_to(map_geojson)

# Agregar control de capas al mapa (permite activar o desactivar la capa)
folium.LayerControl().add_to(map_geojson)

# Mostrar el mapa
map_geojson
```
