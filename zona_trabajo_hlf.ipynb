```python

import numpy as np

class HundirLaFlota:
    def __init__(self ,filas=10, columnas=10):
        """
        Inicializa el juego con el tamaño del tablero especificado.
        """
        self.filas = filas
        self.columnas = columnas
        self.tablero_jugador = np.full((self.filas, self.columnas), " ")
        self.tablero_ordenador = np.full((self.filas, self.columnas), " ")
        self.barcos = barcos=[5, 4, 3, 3, 2]

    def colocar_barcos_aleatorios(self,tablero, n_barcos = 4):
        """
        Coloca los barcos aleatoriamente en el tablero.
        """
        while n_barcos >= 1:
            # Genera posiciones aleatorias en el tablero
            fila = np.random.randint(0, tablero.shape[0])
            columna = np.random.randint(0, tablero.shape[1])
            # Escoge aleatoriamente el tamaño de un barco
            size = np.random.choice(self.barcos)

            # Verifica si el barco puede ser colocado en la posición [fila, columna] sin solaparse con otros barcos
            if self.verificar_adyacencia(tablero, fila, columna, size):
                # Selecciona una orientación aleatoria para el barco
                orientacion = np.random.choice(["horizontal", "vertical"])
                if orientacion == "horizontal":
                    if columna + size <= tablero.shape[1]:
                        tablero[fila, columna:columna + size] = "B"
                        n_barcos -= 1
                else:
                    if fila + size <= tablero.shape[0]:
                        tablero[fila:fila + size, columna] = "B"
                        n_barcos -= 1

    def verificar_adyacencia(self, tablero, fila, columna, size):
        """
        Verifica si hay suficiente espacio alrededor de una posición para colocar un barco sin solaparse con otros barcos.
        """
        for i in range(-1, size + 1):
            for j in range(-1, size + 1):
                if 0 <= fila + i < tablero.shape[0] and 0 <= columna + j < tablero.shape[1]:
                    if tablero[fila + i, columna + j] == "B":
                        return False
        return True
    
    def disparo_simple(self, tablero, fila, columna):
        '''
        En la entrada de esta función hay que hacer referencia al tablero del contrincante.
        '''

        if tablero[fila, columna] == "B":
            tablero[fila, columna] = "X"
            print("Has tocado un barco")
            print(tablero)
        elif tablero[fila, columna] == "X":
            print("Ya has disparado a esta posición")
        else:
            tablero[fila, columna] = "0"
            print("¡Agua!")
            print(tablero)

    def turno_ordenador(self):
        fila = np.random.randint(0, self.tablero_ordenador.shape[0])
        columna = np.random.randint(0, self.tablero_ordenador.shape[1])
        self.disparo_simple(self.tablero_jugador, fila, columna)
        print(f"El ordenador a disparado en la posición {fila}{columna}")

    def jugar(self):
        self.colocar_barcos_aleatorios(self.tablero_ordenador)
        print("Tablero ordenador")
        print(self.tablero_ordenador)
        self.colocar_barcos_aleatorios(self.tablero_jugador)
        print("Tablero jugador")
        print(self.tablero_jugador)

        while True:
            print("\n Tu turno")
            fila = int(input("Fila a disparar?"))
            columna = int(input("Columna a disparar?"))
            self.disparo_simple(self.tablero_ordenador, fila, columna)
            print(self.tablero_jugador)

            if np.all(self.tablero_ordenador == "X"):
                print("Felicidades, has ganado!")
                break

            print("\n Turno del ordenador")
            self.turno_ordenador()

            if np.all(self.tablero_jugador == "X"):
                print("Has perdido, no te quedan más barcos")
                break






```
