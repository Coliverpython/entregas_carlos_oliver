## 1. Create an interactive bar chart showing total sales by State for2014.

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
Df_2014 = df[df["Order Date"].dt.year == 2014]
Df_2014
```

```python
# Filter the DataFrame for the year 2014
df_2014 = df[df['Order Date'].dt.year == 2014]

# Group and aggregate data
sales_by_state = df_2014.groupby('State')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)

# Create interactive bar chart using Plotly
fig = px.bar(sales_by_state, x='State', y='Sales', 
             labels={'Sales': 'Total Sales', 'State': 'State'}, 
             title='Total Sales by State for the Year 2014',
             color='State')  # Color bars by state for better differentiation

# Show the interactive chart
fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Assuming 'df' is your DataFrame after reading from a CSV or other data source

# Filter the DataFrame for the states California, New York, and Texas
df_california = df[df['State'] == 'California']
df_new_york = df[df['State'] == 'New York']
df_texas = df[df['State'] == 'Texas']

# Create a subplot with 1 row and 3 columns
fig = make_subplots(rows=1, cols=3, subplot_titles=('California', 'New York', 'Texas'))

# Add a histogram to each subplot for the corresponding state
fig.add_trace(go.Histogram(x=df_california['Quantity'], name='California'), row=1, col=1)
fig.add_trace(go.Histogram(x=df_new_york['Quantity'], name='New York'), row=1, col=2)
fig.add_trace(go.Histogram(x=df_texas['Quantity'], name='Texas'), row=1, col=3)

# Update the layout of the figure
fig.update_layout(title_text='Quantity per order in 3 States', showlegend=False)
fig.update_xaxes(title_text='Quantity')
fig.update_yaxes(title_text='Count')

# Show the figure
fig.show()
```

## 4. Create an interactive stacked bar chart showing the revenue by state (bars) for each of the products categories for California, New York and Texas.

```python
import plotly.express as px

# Filtrar los datos para incluir solo los estados de California, Nueva York y Texas
filtered_df = df[df['State'].isin(['California', 'New York', 'Texas'])].groupby(['State', 'Category'], as_index= False)[['Sales']].sum()

ventas_por_categoria = filtered_df.groupby('Category')['Sales'].sum().reset_index()


# Crear el gr√°fico de barras interactivo con Plotly Express
fig = px.bar(filtered_df, x='Category', y='Sales', color='State',
             labels={'Sales': 'Total Sales', 'Category': 'Category'},
             title='Total Sales by Category for California, New York, and Texas')

# Mostrar el gr√°fico interactivo
fig.show()
```

```python
import plotly.express as px

# Filtrar los datos para incluir solo los estados de California, Nueva York y Texas
filtered_df = df[df['State'].isin(['California', 'New York', 'Texas'])]

# Sumar las ventas por categor√≠a y ciudad
ventas_por_categoria_y_ciudad = filtered_df.groupby(['Category', 'City'])['Sales'].sum().reset_index()

# Crear el gr√°fico de barras interactivo con Plotly Express
fig = px.bar(ventas_por_categoria_y_ciudad, x='Category', y='Sales',
             facet_row='City', color='City', facet_row_spacing=0.03,
             labels={'Sales': 'Total Sales', 'Category': 'Category'},
             title='Total Sales by Category for California, New York, and Texas (Faceted by City)')

# Mostrar el gr√°fico interactivo
fig.show()


```

## 5. Create an interactive line chart showing sale by day for California.

```python
filtered_df = df[df['Order date'].isin(['California', 'New York', 'Texas'])]
```

```python
import plotly.express as px

# Filtrar los datos para incluir solo las filas correspondientes a California
california_data = df[df['State'] == 'California']

# Agrupar los datos por d√≠a y sumar las ventas para cada d√≠a
sales_by_day = california_data.groupby('Order Date')['Sales'].sum().reset_index()

# Crear el gr√°fico de l√≠nea interactivo con Plotly Express
fig = px.line(sales_by_day, x='Order Date', y='Sales', title='Sales by Day in California')
fig.update_xaxes(title='Date')
fig.update_yaxes(title='Sales')

# Mostrar el gr√°fico interactivo
fig.show()
```
